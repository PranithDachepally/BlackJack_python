```python
import random

suits = ('Hearts','Diamonds','Spades','Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':11}
playing = True


class Deck():
    
    #creating a deck of cards
    def __init__(self):
        self.deck=[]
        for s in suits:
            for r in ranks:
                self.deck.append(r + ' of ' + s)             
    def __str__(self):
        return ''.join(str(e)+'\n' for e in self.deck)

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        card = self.deck.pop()
        return card

class Hand:
    #adding card to hand and adjusting the value of ace's in the hand
    def __init__(self):
        self.cards = []
        self.aces = 0  
        self.value = 0
        
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[str(card.split(" ")[0])]
    
    def adjust_for_ace(self):
        while (self.value > 21 and self.aces>1):
            self.value = self.value-10
            self.aces = self.aces-1
            
class Chips:
    #Player chips class
    def __init__(self):
        self.total = 100
        self.bet = 0
        
    def win_bet(self):
        self.total = self.total + self.bet
    
    def lose_bet(self):
        self.total = self.total - self.bet
    
def take_bet(chips):
    #checks if the bet amount matches the total chips value 
    while True:
        try:
            chips.bet = int(input('enter the betting amount :  '))
        except ValueError:
            print('Sorry enter a integer value')   
        else:
            if(chips.bet > chips.total):
                print('you dont have sufficient chips')
            else:
                break

def hit(desk,hand):
    hand.add_card(desk.deal())
    hand.adjust_for_ace()
        
def hit_or_stand(deck,hand):
    #hit or stand operations deciding funtion
    global playing
    
    while True:
        x = input('hit or stand :  ')
        if(x.lower().startswith('h')):
            hit(deck,hand)         
        elif(x.lower().startswith('s')):  
            print('player stands.Dealer is playing')
            playing = False
        else:
            print('Sorry, try again')
            continue
        break

def show_some(player,dealer):
    #shows all the cards of player and dealer excluding one hidden card
    print('\nshowing cards')
    print('player cards: '+str(player.cards))
    print('dealer cards: '+str(dealer.cards[1])) 
    
def show_all(player,dealer):
    #shows all the cards of player and dealer
    print('\nshowing cards')
    print('player cards: '+str(player.cards))
    print('player value : '+str(player.value))
    print('dealer cards: '+str(dealer.cards)) 
    print('dealer value : '+str(dealer.value))
    
def player_busts(player,dealer,chips):
    print('Player Bust!!!')
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print('Player won!')
    chips.win_bet()
    
def dealer_busts(player,dealer,chips):
    print('Dealer Bust!!!')
    chips.win_bet()
    
def dealer_wins(player,dealer,chips):
    print('Dealer won!')
    chips.lose_bet()
    
def push(player,dealer):
    print('Draw!')    
    
while True:
    
    print('Welcome to BlackJack!')
    
    #Creating & shuffling the deck, deal two cards to each player
    deck = Deck()
    deck.shuffle()
    
    #Creating a player hand
    player = Hand()
    player.add_card(deck.deal())
    player.add_card(deck.deal())
    
    #Creating a dealer hand
    dealer = Hand()
    dealer.add_card(deck.deal())
    dealer.add_card(deck.deal())
    
    #Setting player's chips
    player_chips = Chips()
    
    #Betting
    take_bet(player_chips)
    
    #Showing cards of player and dealer excluding the hidden card of the dealer
    show_some(player,dealer)
    
    while playing:
        
        #Asks the player to hit or stand
        hit_or_stand(deck,player)
        
        #Showing cards of player and dealer excluding the hidden card of the dealer
        show_some(player,dealer)
        
        #Checking player cards value
        if player.value > 21:
            player_busts(player,dealer,player_chips)
            break
    
    #Dealer's turn    
    if player.value <= 21:
        
        while dealer.value < 17 :
            hit(deck,dealer)
        #Showing all the cards of player and dealer
        show_all(player,dealer)
        
        #Checking for winner
        if dealer.value > 21 :
            dealer_busts(player,dealer,player_chips)
        elif dealer.value > player.value:
            dealer_wins(player,dealer,player_chips)
        elif dealer.value < player.value :
            player_wins(player,dealer,player_chips)
        else:
            push(player,dealer)
    
    #Printing total number of remaining chips
    print('Player chips : ',player_chips.total)
    
    #Asking for playing again
    play_again = input('Would you like to play again(yes/no) : ')
    
    if (play_again.lower().startswith('y')):
        playing = True
        continue
    else:
        print('Thanks for playing!')
        break
    
```
